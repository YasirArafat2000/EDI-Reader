```python
import pandas as pd
import numpy as np
```

```python
with open(text_name, "r") as f:
    content = f.readlines()
modified2 = []
for line in content :
    if line[-1] =="\n":
        modified2.append(line[:-1])
    else:
        modified2.append(line)
print(modified2)
```

```python
modified3 = []
for line in modified2:
    if line.__contains__('XQ'):
        modified3.append(line)
    elif line.__contains__('LIN'):
        modified3.append(line)
    elif line.__contains__('SDQ'):
        modified3.append(line)
    else:
        print(line)

```

```python
print(modified3)
```

```python
characters = []
for word in modified3:
    if word not in characters:
        characters.append(word.split("*"))

edi2 = pd.DataFrame(characters)
edi2.columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
```

```python
edi3 = edi2.fillna(0)
edi3.head(20)
```

```python
# copy down the values in "new_column" if "E" is empty
###df['new_column'] = df['E'].where(df['E'].str.len() > 0, df['V'].shift(1))
```

```python

# Create a new column called SKU ID and initialize it to 0
edi3['SKU ID'] = 0

# Iterate over each row of the DataFrame
last_sku_id = 0
for i, row in edi3.iterrows():
    # Check if the value in column H is not null
    if pd.notnull(row['H']):
        # If it's not null, assign the value to the SKU ID column
        sku_id = row['H']
        edi3.at[i, 'SKU ID'] = sku_id
        last_sku_id = sku_id
    else:
        # If it's null, assign the last non-null SKU ID value to the current row
        edi3.at[i, 'SKU ID'] = last_sku_id
```

```python
edi3['SKU ID'] = edi3['SKU ID'].astype(str)
edi3.replace({'SKU ID': {'0': np.nan}}, inplace=True)
```

```python
edi3['SKU ID'].fillna(method='ffill', inplace=True)
edi3.fillna(0, inplace=True)
```

```python
edi3.head()
```

```python
# rename columns D and E
edi3.rename(columns={'D': 'Store ID', 'E': 'Quantity'}, inplace=True)

# fill missing values in SKU ID column
edi3['SKU ID'].fillna(method='ffill', inplace=True)

# replace "UP" values with NaN values in Quantity column
edi3['Quantity'].replace('UP', pd.np.nan, inplace=True)

# copy down the value in the first row of column C
if edi3.loc[edi3['C'] != '', 'C'].empty:
    edi3.loc[:, 'C'] = edi3.iloc[0]['C']
else:
    edi3.loc[edi3['C'] == '', 'C'] = edi3.loc[edi3['C'] != '', 'C'].iloc[0]

# forward-fill blank values in column C
edi3['C'].fillna(method='ffill', inplace=True)

# convert Quantity column to numeric type
edi3['Quantity'] = pd.to_numeric(edi3['Quantity'])

# create Quantity and Quantity Returned columns
edi3['Quantity'] = edi3['Quantity'].apply(lambda x: x if x >= 0 else 0)
edi3['Quantity Returned'] = edi3['Quantity'].apply(lambda x: abs(x) if x < 0 else 0)



# reorganize columns
edi3 = edi3[['Store ID', 'SKU ID', 'Quantity', 'Quantity Returned', 'A', 'B', 'C', 'F', 'G', 'H']]

```

```python
replacement = { 'BP': edi3['C'].iloc[0] }
edi3['C'] = edi3['C'].replace(replacement)

# convert column C to datetime
edi3['C'] = pd.to_datetime(edi3['C'], format='%Y%m%d')

# add a new column with 6 days added to column C
edi3['Ending Date'] = edi3['C'] + pd.Timedelta(days=6)

# rename column C to Beginning Date
edi3.rename(columns={'C': 'Beginning Date'}, inplace=True)

# reorder the columns
edi3 = edi3.reindex(columns=['Beginning Date', 'Ending Date', 'Store ID','SKU ID', 'Quantity', 'Quantity Returned', 'A', 'B', 'F', 'G', 'H'])
```
